```python
cat = ['alt.atheism' , 'comp.graphics' ,'sci.space']
train = fetch_20newsgroups(subset = 'train' , categories = cat)
test = fetch_20newsgroups(subset = 'test' , categories = cat)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

cnt = CountVectorizer(max_features = 1000)
x_train = cnt.fit_transform(train.data)
```

```python
cnt.get_feature_names_out()
```

```python
from sklearn.naive_bayes import BernoulliNB , MultinomialNB

bnb = BernoulliNB()
bnb.fit(x_train,train.target)
```

```python
x_test = cnt.transform(test.data)
```

```python
y_pred = bnb.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(test.target , y_pred)
acc
```

```python
tf = TfidfVectorizer(max_features = 1000 , stop_words='english')
x1_train = tf.fit_transform(train.data)
```

```python
mnb = MultinomialNB()
mnb.fit(x1_train , train.target)
```

```python
x1_test = tf.transform(test.data)
```

```python
y_pred = mnb.predict(x1_test)
y_pred
```

```python
acc = accuracy_score(test.target , y_pred)
acc
```

```python
# Remove stopwords in CountVectorizer or TfidfVectorizer
vectorizer_stop = CountVectorizer(stop_words='english')
X_train_stop = vectorizer_stop.fit_transform(train.data)
X_test_stop = vectorizer_stop.transform(test.data)

model3 = BernoulliNB()
model3.fit(X_train_stop, train.target)

y_pred_stop = model3.predict(X_test_stop)
accuracy_stop = accuracy_score(test.target, y_pred_stop)
print("Accuracy without stopwords:", accuracy_stop)
```
